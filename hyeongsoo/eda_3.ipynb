## 0) Original Image

```python
#   
image = cv2.imread("./data/train/00ec594b7cc89183.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image), plt.axis("off")
plt.show()
```

## 1) BrightnessTexturize

```python
brightness_texturize = BrightnessTexturize(texturize_range=(0.99, 0.99),
                                           deviation=0.1 )

img_brightness_texturize = brightness_texturize(image)
plt.imshow(img_brightness_texturize), plt.axis("off")
plt.show()
```

## 2) InkBleed

```python
inkbleed = InkBleed(intensity_range=(0.4, 0.4),
                    kernel_size=(5, 5),
                    severity=(0.4, 0.4)
                    )

img_inkbleed = inkbleed(image)
plt.imshow(img_inkbleed), plt.axis("off")
plt.show()
```

## 3) LowLightNoise

```python
lowlightnoise_obj = LowLightNoise(
    num_photons_range = (88, 88),
    alpha_range = (0.7, 0.7),
    beta_range = (10, 10),
    gamma_range = (1.0 , 1.0)
)

lowlightnoise_img = lowlightnoise_obj(image)
plt.imshow(lowlightnoise_img), plt.axis("off")
plt.show()
```

## 4) NoiseTexturize

```python
noise_texturize = NoiseTexturize(sigma_range=(3, 3),
                                 turbulence_range=(3, 3),
                                 texture_width_range=(300, 300),
                                 texture_height_range=(300, 300),
                                 )

img_noise_texturize = noise_texturize(image)
plt.imshow(img_noise_texturize), plt.axis("off")
plt.show()
```

## 5) SubtleNoise

```python
subtle_noise = SubtleNoise(subtle_range=25)

img_subtle_noise = subtle_noise(image)
plt.imshow(img_subtle_noise), plt.axis("off")
plt.show()
```
